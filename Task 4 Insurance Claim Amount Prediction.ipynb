```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Random seed for reproducibility
np.random.seed(100)

# Number of records
num_records = 200

# Creating sample customer data
age = np.random.randint(18, 65, num_records)
bmi = np.random.uniform(18, 35, num_records)
smoking = np.random.choice([0, 1], size=num_records, p=[0.75, 0.25])

# Generating insurance charges
charges = (
    age * 180 +
    bmi * 280 +
    smoking * 13000 +
    np.random.normal(0, 1000, num_records)
)

# Creating DataFrame
insurance_data = pd.DataFrame({
    "Age": age,
    "BMI": bmi,
    "Smoking_Status": smoking,
    "Insurance_Charges": charges
})

print("Dataset Preview")
print(insurance_data.head())

print("\nDataset Shape:", insurance_data.shape)

# BMI vs Insurance Charges
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=insurance_data,
    x="BMI",
    y="Insurance_Charges",
    hue="Smoking_Status"
)

plt.title("Relationship Between BMI and Insurance Charges")
plt.xlabel("BMI")
plt.ylabel("Insurance Charges")
plt.show()
```
